```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tunes = pd.read_csv('./results_generalRe/results.csv')
channels = tunes[['Out_C', 'Val Acc', 'Val Acc2']]
decays = tunes[['Weight D', 'Val Acc.1', 'Val Acc2.1']]
decays = decays.rename(columns={'Val Acc.1': 'Val Acc', 'Val Acc2.1': 'Val Acc2'})
decays = decays.loc[decays['Weight D'].notna(), :]
```

```python
fig, ax = plt.subplots(1, 2, figsize=(9, 4.5))
ax = ax.flatten()
ax[0].plot(channels['Out_C'], channels['Val Acc'], '-', label='Undersampling')
ax[0].plot(channels['Out_C'], channels['Val Acc2'], '-', label='Oversampling')
ax[0].set_ylim([0.675, 0.755])
ax[0].set_xlabel('Number of output channels')
ax[0].set_ylabel('Validation accuracy')
ax[0].legend()
ax[1].plot(decays['Weight D'], decays['Val Acc'], '-', label='Undersampling')
ax[1].plot(decays['Weight D'], decays['Val Acc2'], '-', label='Oversampling')
ax[1].set_xlim([0.0038, 0.0162])
ax[1].set_ylim([0.675, 0.755])
ax[1].set_xlabel('Weight decay')
ax[1].set_ylabel('Validation accuracy')
ax[1].legend()
plt.tight_layout()
plt.rcParams.update({'font.size': 12})
plt.savefig('./results_generalRe/cnn_tune_plots.png', dpi=500, bbox_inches='tight')
```
